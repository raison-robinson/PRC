```python
#load the data excel sheet
import pandas as pd
file=pd.read_csv("//content/irisexcel.csv")
x=file[["sepal_length","sepal_width","petal_length","petal_width"]]
y=file["species"]
#cross validation
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.75)
#load the model
from sklearn.linear_model import Perceptron
model=Perceptron()
#fit the model
model.fit(xtrain,ytrain)
#validation
op=model.predict(xtest)
from sklearn.metrics import accuracy_score
score=accuracy_score(ytest,op)
print("Accuracy=",score*100)
#testing
op2=model.predict([[4.9,3,1.2,0.2]])
print(op2)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm=confusion_matrix(ytest,op)
print(cm)
cmd=ConfusionMatrixDisplay(confusion_matrix=cm,

                              display_labels=model.classes_)
cmd.plot()
```
